Распределение разных колонок относительно HomePlanet примерно одинаковое, зависимости не выявлено, поэтому заменим пропущенные значения на Earth, так как их большинство. И заменим все Europe на Earth.

```python
train_test_data = [train, test] # combining train and test dataset
# Earth = 0, Mars = 1
for dataset in train_test_data:
  dataset['HomePlanet'] = dataset['HomePlanet'].fillna('Earth')

train.head()
```
